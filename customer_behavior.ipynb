```python
import pandas as pd
df=pd.read_csv("customer_shopping_behavior.csv")
df.head()
```

```python
df.info()
```

```python
df.describe(include="all")
```

```python
df.isnull().sum()
```

```python
df["Review Rating"]=df.groupby("Category")["Review Rating"].transform(lambda x:x.fillna(x.median()))
```

```python
df.isnull().sum()
```

```python
df.columns= df.columns.str.lower()
df.columns=df.columns.str.replace(" ","_")
df = df.rename(columns={'purchase_amount_(usd)':'purchase_amount'})
df
```

```python
# Create a column of age_group
```

```python
labels=["Young Adult","Adult","Middle Age","Senior"]
df["age_group"]=pd.qcut(df["age"],q=4,labels=labels)
```

```python
df[["age_group"]].head(10)
```

```python
# create  a column purchase_frequecy_days
```

```python
frequency_mapping={
    "Fortnightly":14,
    "Weekly":7,
    "Monthly":30,
    "Quarterly":90,
    "Bi-Weekly":14,
    "Annually":365,
    "Every 3 Months":90
}
df["purchase_frequecy_days"]=df["frequency_of_purchases"].map(frequency_mapping)
df[["purchase_frequecy_days","frequency_of_purchases"]].head(10)
```

```python
df[["discount_applied","promo_code_used"]].head(10)
```
